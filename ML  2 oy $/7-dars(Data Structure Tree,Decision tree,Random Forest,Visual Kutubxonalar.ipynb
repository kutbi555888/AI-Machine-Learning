# Tree in models

```python
import pandas as pd
df=pd.read_csv('diabetes_dataset.csv')
```

```python
df.info()
```

```python
from datacleaner import autoclean
df=autoclean(df)
```

```python
df.info()
```

```python
x=df.drop('Family_History_of_Diabetes',axis=1)
```

```python
x
```

```python
y=df['Family_History_of_Diabetes']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier(max_depth=3,criterion='gini')
```

```python
model
```

```python
model.fit(x_train,y_train)
```

```python
from sklearn.tree import plot_tree       # treeni orqa tomonda poissonlar bilan qanday ishlayotganini korsatib beradi
import matplotlib.pyplot as plt    # bu visual korinish uchun
```

```python
plt.figure(figsize=(20,10))      
plot_tree(model,feature_names=x.columns,class_names=['Prediabetic','Non-prediabetic'],filled=True)
plt.show()
```

# Random Forest

```python
import pandas as pd
df=pd.read_csv('diabetes_dataset.csv')
```

```python
df.info()
```

```python
from datacleaner import autoclean
df=autoclean(df)
```

```python
df.info()
```

```python
x=df.drop('Family_History_of_Diabetes',axis=1)
```

```python
x
```

```python
y=df['Family_History_of_Diabetes']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RFmodel=RandomForestClassifier(max_depth=3,criterion='gini')
```

```python
RFmodel
```

```python
RFmodel.fit(x_train,y_train)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,10))      
plot_tree(RFmodel.estimators_[0],feature_names=x.columns,class_names=['Prediabetic','Non-prediabetic'],filled=True)
plt.show()
# RFmodel.estimators_ — bu Random Forest ichidagi barcha DecisionTreeClassifier obyektlarining ro‘yxati.
# estimators_[0] — birinchi daraxt.
# Random Forest kop daraxtdan iborat model.lekin plot_tree faqat 1ta daraxtni chizadi.shuning uchun 0 bu 1chi daraxt.yani faqat bitta daraxtni korsa boladi

```
