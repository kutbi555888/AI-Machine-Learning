```python
import pandas as pd
```

```python
df=pd.read_csv('top_expensive_leagues.csv')
```

```python
df.info()
```

```python
from datacleaner import autoclean
```

```python
df=autoclean(df)
```

```python
df.info()
```

```python
x=df.drop('Viewership',axis=1)
```

```python
x
```

```python
y=df['Viewership']
```

```python
from sklearn.model_selection import train_test_split,KFold, cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.tree import DecisionTreeRegressor
```

```python
x_train,x_temp,y_train,y_temp=train_test_split(x,y,test_size=0.2,random_state=42)
x_test,x_val,y_test,y_val=train_test_split(x_temp,y_temp,test_size=0.5,random_state=42)
```

```python
model=DecisionTreeRegressor()
nonlinear_model=model.fit(x_train,y_train)
y_pred=nonlinear_model.predict(x_test)
mse=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
```

```python
model.fit(x_train,y_train)
```

```python
mse
```

```python
r2
```

```python
kf = KFold(n_splits=10, shuffle=True, random_state=42)
```

```python
print(kf)
```

```python
scores=cross_val_score(model,x_train,y_train,cv=kf,scoring='r2')
```

```python
scores
```

```python
scores.mean()
```

```python
scores.std()
```

```python
# grid search parametr
param_grid={
    'max_depth' :[3,5,10,None],
    'min_samples_split' :[2,5,10],
    'min_samples_leaf' :[1,2,5],
    'max_features' :['auto','sqrt','log','None']
}
```

```python
param_grid
```

```python
from sklearn.model_selection import GridSearchCV
grid_search=GridSearchCV(estimator=model,param_grid=param_grid,
                         cv=5, scoring='r2', n_jobs=-1)
# Fit the model
grid_search.fit(x_train,y_train)
```

```python
print('Best parametrs:',grid_search.best_params_)
```

```python
best_model=grid_search.best_estimator_
```

```python
best_model
```

```python
y_pred=best_model.predict(x_test)
```

```python
r2=r2_score(y_test,y_pred)
```
