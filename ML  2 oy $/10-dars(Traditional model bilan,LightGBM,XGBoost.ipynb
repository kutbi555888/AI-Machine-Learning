# LightGbm mavzusi.1chi Traditional modellar yordamida

```python
import pandas as pd
df=pd.read_csv('global_cancer_patients_2015_2024(in)Classification.csv')
```

```python
df.info()
```

```python
from sklearn.linear_model import LogisticRegression     # Base Line model(algoritm) - Data preprocessingsiz model qilish
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# bu yerda bizni targetimiz regression,shuni uchun target qiymatni toifalarga bo'lib olamiz
# sababi biz classification model ishlatmoqchimiz
# bu yerda biz Target_Severity_Score ustunini ozini toifalarga ajratsak boladi lekin bunday qilsak Target_Severity_Score ustuni ozini asl
# uzluksiz qiymatlarini yoqatadi.keyin regression uchun foydalana olmaymiz.shuning uchun Target_Severity_Score ustunini orniga boshqa toifalarga
# ajratilgan ustun yaratib olamiz.
# bu yerda biz datani regressiondan classificationga otkazyabmiz ana shu flexibellik(moslashuvchanlik)dir.
df['Severity_Class'] = pd.qcut(df['Target_Severity_Score'], q=3, labels=[0, 1, 2])
```

```python
df.info()
```

```python
x=df.drop(['Target_Severity_Score','Severity_Class'],axis=1)
```

```python
x
```

```python
y=df['Severity_Class']
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
lr_model=LogisticRegression()    # classification model
```

```python
lr_model.fit(x_train,y_train)         # sababi bu yerda missing value,encoding qilinmadi.bularni qilmasa model ishlamaydi
```

```python
# endi encoding qilib ko'ramiz
from sklearn.preprocessing import LabelEncoder
```

```python
lb_encoder=LabelEncoder()
for col in df.columns:
    if df[col].dtype=='object':
        if df[col].nunique()<=5:
            dummies=pd.get_dummies(df[col],prefix=col,dtype=int)
            df=pd.concat([df.drop(col,axis=1),dummies],axis=1)
        else:   
                df[col]=lb_encoder.fit_transform(df[col])
```

```python
df.info()                      # bu yerda tushirib qoldirilgan qiymatlari bor lekin encoding boldi.endi x va y ga ajratib fit qilib koramiz
```

```python
lr_model.fit(x_train,y_train)    # endi bu yerda tushirib qoldirilgan qiymatlari bor deyapti    
```

```python
# endi tushirib qoldirilgan qiymatlarni toldiramiz
for col in df.columns:
    if df[col].isnull().any():       # buni manosi agar tushirib qoldirilgan qiymatlar bolsa degani
        if df[col].dtype.name in['object','category']:   # data typi object va categoryni bolganni
            df[col].fillna(df[col].mode()[0],inplace=True)       # mode bilan
        else:
            df[col].fillna(df[col].mean(),inplace=True)         # aks holda numerical qiymat bolsa mean bilan toldir
df['Severity_Class'] = df['Severity_Class'].astype(int)        # bu dtypni ozgartiradi.bu ustun category,shuning uchun Severity_Class ni
# int ga ozgartiradi

```

```python
df.info()    # endi tushirib qoldirilgan qiymatlar ham tozalandi,encoding ham qilindi,Severity_Class int ga ozgardi
```

```python
# endi x va y ga ajratib fit train qilamiz
lr_model.fit(x_train,y_train)
```

```python
y_pred=lr_model.predict(x_test)
```

```python
score1=accuracy_score(y_test,y_pred)
```

```python
score1
```

```python
# bu yerda bizni targetimiz regression,shuni uchun target qiymatni toifalarga bo'lib olamiz
# sababi biz classification model ishlatmoqchimiz
# bu yerda biz Target_Severity_Score ustunini ozini toifalarga ajratsak boladi lekin bunday qilsak Target_Severity_Score ustuni ozini asl
# uzluksiz qiymatlarini yoqatadi.keyin regression uchun foydalana olmaymiz.shuning uchun Target_Severity_Score ustunini orniga boshqa toifalarga
# ajratilgan ustun yaratib olamiz.
# bu yerda biz datani regressiondan classificationga otkazyabmiz ana shu flexibellik(moslashuvchanlik)dir.
df['Severity_Class'] = pd.qcut(df['Target_Severity_Score'], q=3, labels=[0, 1, 2])
```

```python
df.info()
```

```python
x=df.drop(['Target_Severity_Score','Severity_Class'],axis=1)
```

```python
y=df['Severity_Class']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
lgbmodel=lgb.LGBMClassifier()
```

```python
lgbmodel
```

```python
lgbmodel.fit(x_train,y_train)      # shu yerda Tushib qolgan qiymatlar bor deyapti.lekin LightGbm model Nan qiymat bolsa ham ishlashi kerakku
```

```python
# Tushib qolgan qiymatlarni toldirib koramiz
for col in df.columns:
    if df[col].isnull().any():    # bu int va float uchun     
        if df[col].dtype=='category':   # bu category uchun
            df[col].fillna(df[col].mode()[0],inplace=True)       
        else:
            df[col].fillna(df[col].mean(),inplace=True)       
```

```python
df.info()
```
