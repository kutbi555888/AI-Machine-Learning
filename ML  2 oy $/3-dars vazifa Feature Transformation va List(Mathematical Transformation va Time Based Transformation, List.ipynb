# Time Based Transformation

```python
import pandas as pd
```

```python
df=pd.read_csv('IMDB_Movies_Dataset.csv')
```

```python
df.info()
```

```python
df['Release Date']
```

```python
df[['Date', 'Country']] = df['Release Date'].str.extract(r'(.+?)\s*\((.+)\)')
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month_name()
df['Day'] = df['Date'].dt.day
```

```python
df.info()
```

```python
df.drop(columns=['Date', 'Unnamed: 0'], inplace=True)
```

```python
df.info()
```

```python
df.drop('Release Date', axis=1, inplace=True)
```

```python
df.info()
```
