# Mutual info

```python
import pandas as pd
df=pd.read_csv('wine_quality_classification.csv')
```

```python
df.info()
```

```python
df['quality_label']
```

```python
df.isnull().sum()
```

```python
# Encoding
from sklearn.preprocessing import LabelEncoder
```

```python
lb_encoder=LabelEncoder()
```

```python
df['quality_label_class']=lb_encoder.fit_transform(df['quality_label'])
```

```python
df['quality_label_class']
```

```python
df.info()
```

```python
df.drop('quality_label',axis=1,inplace=True)   # df dagi ustunni drop qiladi
```

```python
df.info()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
# y qiymatdan boshqa barcha ustunlarni scaling qiladi
for col in df.columns:
    if col!='quality_label_class':
        df[col]=scaler.fit_transform(df[[col]])
```

```python
df.info()
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop('quality_label_class',axis=1)
```

```python
x
```

```python
y=df['quality_label_class']
```

```python
y
```

```python
x_train,x_temp,y_train,y_temp=train_test_split(x,y,test_size=0.2,random_state=42)
x_test,x_val,y_test,y_val=train_test_split(x_temp,y_temp,test_size=0.5,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr_model=LogisticRegression()
```

```python
lr_model
```

```python
lr_model.fit(x_train,y_train)
```

```python
y_pred=lr_model.predict(x_val)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc=accuracy_score(y_val,y_test)
```

```python
acc
```

```python
# MUTUAL INFO
from sklearn.feature_selection import mutual_info_classif
```

```python
mi_scores1=mutual_info_classif(x,y)
```

```python
mi_scores1
```

```python
mi_scores1_df=pd.DataFrame(mi_scores1,index=x.columns,columns=['Featurelarning Target qiymatni predict qilishda qoshayotgan hissasi'])
```
