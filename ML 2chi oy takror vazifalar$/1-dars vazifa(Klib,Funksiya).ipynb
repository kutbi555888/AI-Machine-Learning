# Klib,Logistic Regression,Funksiya

```python
import pandas as pd
df=pd.read_csv('tuberculosis_xray_dataset 4Classification.csv')
```

```python
df.info()
```

```python
import klib
```

```python
df=klib.data_cleaning(df)
```

```python
df.info()
```

```python
df=df.drop_duplicates()      # pandas yordamida hammasini duplicated qilish
```

```python
# Missing value
df['gender'].fillna(df['gender'].mode()[0],inplace=True)
df['cough_severity'].fillna(df['cough_severity'].mean(),inplace=True)
df['smoking_history'].fillna(df['smoking_history'].mode()[0],inplace=True)
df['previous_tb_history'].fillna(df['previous_tb_history'].mode()[0],inplace=True)
df['class'].fillna(df['class'].mode()[0],inplace=True)
```

```python
df.info()
```

```python
# For loop orqali missing value
for col in df.columns:
    if df[col].isnull().any():
        if df[col].dtype=='category':
            df[col].fillna(df[col].mode()[0],inplace=True)
        else:
            df[col].fillna(df[col].mean(),inplace=Tue)
```

```python
# Encoding
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
```

```python
encoder
```

```python
no_numeric_col=df.select_dtypes(include=['string','category']).columns
```

```python
no_numeric_col
```

```python
df[no_numeric_col]=encoder.fit_transform(df[no_numeric_col])      # Sababi bu umumiy har bittasiga kirib chiqib encoding qila olmaydi.
# faqat 1ta 1talab ustunni encoding qilmoqchi bolsak ishlaydi
```

```python
# masalan 1ta ustunni shu usul bilan encoding qilib koramiz
df['age']=encoder.fit_transform(df['age'])
# bitta ustunni encoding qildi
```

```python
# endi biz hammasini encoding qilmoqchi bolsak har bittasiga kirib chiqadigan for loop orqali encoding qilishimiz kerak
for col in df[no_numeric_col]:
    df[col]=encoder.fit_transform(df[col])
```

```python
df.info()
```

```python
df['class']
```

```python
x=df.drop('class',axis=1)
```

```python
x
```

```python
y=df['class']
```

```python
y
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_temp,y_train,y_temp=train_test_split(x,y,test_size=0.2,random_state=42)
x_test,x_val,y_test,y_val=train_test_split(x_temp,y_temp,test_size=0.5,random_state=42)
```

```python
x_test.shape
```

```python
x_val.shape
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_val)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
score=accuracy_score(y_test,y_pred)    # qiymmatlar tengligi uchun y_test ishladi
```
