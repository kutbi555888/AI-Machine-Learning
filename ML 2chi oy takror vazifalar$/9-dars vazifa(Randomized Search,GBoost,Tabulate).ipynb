```python
import pandas as pd
df=pd.read_csv('personality_dataset.csv')
```

```python
df.info()
```

```python
def tozala(df):
    for col in df.columns:
        if df[col].isnull().any():
            if df[col].dtype=='object':
                df[col].fillna(df[col].mode()[0],inplace=True)
            else:
                df[col].fillna(df[col].mean(),inplace=True)

    return df.info()
```

```python
tozala(df)
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

encoder = LabelEncoder()
```

```python
def encodlash(df):
    df = df.copy()
    target_col = 'Personality'    # bu yerda target ustunni belgilab olyapmiz
    encoder = LabelEncoder()

    #  Biz bu yerda Target ustunni Label Encoding bilan qilyapmiz
    if df[target_col].dtype == 'object':
        df[target_col] = encoder.fit_transform(df[target_col])
    
    # Boshqa object ustunlarni kodlash
    for col in df.columns:
        if col == target_col:
            continue  # target ustunni kodlab boâ€˜ldik
        if df[col].dtype == 'object':
            if df[col].nunique() <= 6:
                dummies = pd.get_dummies(df[col], prefix=col, dtype=int)
                df = pd.concat([df.drop(columns=[col]), dummies], axis=1)
            else:
                df[col] = encoder.fit_transform(df[col])

    return df
```

```python
df=encodlash(df)
```

```python
df.info()
```

```python
# Random Forestga scaling kerak emas.lekin kodini yozib qoyamiz
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
```

```python
def scaling(df,target_col):
    df=df.copy()    
    numeric_col=df.select_dtypes(include=['number']).columns.drop(target_col)
   
    df[numeric_col]=scaler.fit_transform(df[numeric_col])
    return df
```

```python
df=scaling(df,target_col="Personality")     # Personality objectligi uchun uni 1chi raqamliga o'tkazib olish kerak
```

# GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'n_estimators': [50, 100, 150],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5]
}


grid_search = GridSearchCV(
    estimator=rf_model,
    param_grid=param_grid,
    cv=3,              
    scoring='accuracy',
    n_jobs=-1,         
    verbose=2
)
```

```python
grid_search.fit(x_train, y_train)
```

```python
print("Best parameters:", grid_search.best_params_)
```

```python
print("Best accuracy:", grid_search.best_score_)
```
