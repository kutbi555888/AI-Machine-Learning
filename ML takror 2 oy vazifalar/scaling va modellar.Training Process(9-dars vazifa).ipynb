# Data Preprocessing

```python
from datacleaner import autoclean
```

```python
df=autoclean(df)
```

```python
df.info()
```

```python
df.head()
```

```python
#Scaling
```

```python
num_col=df.select_dtypes(include='number').columns
```

```python
#1 Min- Max scaling
scaler_min_max=MinMaxScaler()
df[num_col]=scaler_min_max.fit_transform(df[num_col])
```

```python
df.head()
```

```python
#2 Standart scaling
scaler_standart=StandardScaler()
df[num_col]=scaler_standart.fit_transform(df[num_col])
```

```python
df.head()
```

```python
#3 Robust scaling
scaler_robust=RobustScaler()
df[num_col]=scaler_robust.fit_transform(df[num_col])
```
