```python
# Robotni turgan  positisiyasi  berilgan va robotni ishlash qoidasiga kora chapdan onga xarakatlanadi va berilgan buyruq songa ko'ra 3 marta xarakatlanadi va bir marta orqaga xarakatlanadi. Robotni yangi joyini topadigan funksiya yarating:
# Masalan  start=0 buyruq=2 natija 4
def robot_yangi_joy(start, buyruq):
    return start + buyruq * 2
```

```python
print(robot_yangi_joy(0, 2))
print(robot_yangi_joy(5, 3))
print(robot_yangi_joy(-2, 4))
```

```python
#2 Berilgan  2 ta kubnining orqasidan eng kattasini toping(1 qator kod orqali)
#  har bir sonning kubini hisoblab, keyin ularning eng kattasini topish
# 1 qatorli funksiya (lambda uslubida):
def eng_katta_kub(a, b): return max(a**3, b**3)
```

```python
print(eng_katta_kub(2, 4))
print(eng_katta_kub(5, 3))
```

```python
#4 Berilgan  2 ta kubning  hajmi orasidagi farqni toping

def kub_hajm_farqi(a, b):
    return abs(a**3 - b**3)

# abs(...)	Absolyut qiymat qaytaradi (manfiy bolmasligi uchun musbat korinishida chiqarib beradi sonni)
```

```python
print(kub_hajm_farqi(3, 5))   # 125 - 27 = 98
print(kub_hajm_farqi(6, 2))   # 216 - 8 = 208
```

```python
import pandas as pd
df=pd.read_csv('student_depression_dataset(in).csv')
```

```python
df.info()
```

```python
def tozala(df):
    for col in df.columns:
        if df[col].isnull().any():
            if df[col].dtype=='object':
                df[col].fillna(df[col].mode()[0],inplace=True)
            else:
                df[col].fillna(df[col].mean(),inplace=True)
```

```python
tozala(df)
```

```python
df.info()
```

```python
#Encoding
# Bu faqat misol uchun bu yerda ustunlar ko'pligi uchun One hot bilan emas Label encoding bilan encoding qilishimiz kerak
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
def encod_qil(df):
    df=df.copy()
    
    for col in df.columns:
        if df[col].dtype=='object':
            if df[col].nunique()<=6:
                dummies=pd.get_dummies(df[col],prefix=col,dtype=int)
                df=pd.concat([df.drop(columns=[col]),dummies],axis=1)
            else:
                df[col]=encoder.fit_transform(df[col])
    return df



# Agar quyidagi funksiyada return df qatori bolmasa, unda funksiya hech qanday qiymat qaytarmaydi.
# Keyin bu funksiya chaqirilsa:
# df = encod_qil(df)
# Nima boladi?
# Python avtomatik ravishda None ni qaytaradi.
# df ning qiymati None boladi.

# Agar return df yozilmasa:
# Funksiya hech narsa qaytarmaydi.
# df=encod_qil(df) chaqiruvining natijasi None boladi.
# Asl df esa umuman ozgarmaydi, chunki df.copy() bilan nusxa olingan, va bu nusxa ozgartirilgan, lekin hech qayerga qaytarilmagan.

```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

def encod_qil(df):
    df = df.copy()

    for col in df.columns:
        if df[col].dtype == 'object':
            df[col] = encoder.fit_transform(df[col])
    
    return df
```
