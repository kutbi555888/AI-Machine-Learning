# KNN K-cross validation bilan

```python
import pandas as pd
df=pd.read_csv('student_depression_dataset(in).csv')
```

```python
df.info()
```

```python
df.head()
```

```python
def tozala(df):
    for col in df.columns:
        if df[col].isnull().any():
            if df[col].dtype=='object':
                df[col].fillna(df[col].mode()[0],inplace=True)
            else:
                df[col].fillna(df[col].mean(),inplace=True)
    return df
```

```python
tozala(df)
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
encoder1=LabelEncoder()
```

```python
def encod_qil(df):
    df = df.copy()

    for col in df.columns:
        if df[col].dtype == 'object':
            df[col] = encoder1.fit_transform(df[col])
    
    return df
```

```python
df=encod_qil(df)
```

```python
df.info()
```

```python
df['Depression'] = df['Depression'].astype(int)
```

```python
from sklearn.preprocessing import StandardScaler
scaling1=StandardScaler()
```

```python
def scaling_qil(df):
    num_col=df.select_dtypes(include=['number']).columns.drop('Depression')
    df[num_col]=scaling1.fit_transform(df[num_col])
    return df
```

```python
df=scaling_qil(df)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
x=df.drop('Depression',axis=1)
```

```python
y=df['Depression']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.neighbors import KNeighborsClassifier   # Classification uchun
from sklearn.neighbors import KNeighborsRegressor    # Regression uchun
```

```python
model1=KNeighborsClassifier(n_neighbors=5)
```

```python
model1
```

```python
model1.fit(x_train,y_train)
```

```python
y_pred=model1.predict(x_test)
```

```python
y_pred
```

```python
score1=accuracy_score(y_test,y_pred)
```

```python
print(f"KNN aniqligi:{score1*100}%ga teng")
```

```python
# K- cross validation

from sklearn.model_selection import cross_val_score,KFold
```

```python
kf=KFold(n_splits=5,shuffle=True,random_state=42)
scores1=cross_val_score(model1,x,y,cv=kf,scoring='accuracy')
print(scores1)
```

```python
import numpy as np
```

```python
print(np.mean(scores1).round(2))
# .round(2) -   bu kodda natijada nechta sonni kormoqchi bolsak qavsni ichiga shu sonni yozishimiz mumkin
```

```python
print(np.mean(scores1))
```

```python
np.std(scores1)
```
