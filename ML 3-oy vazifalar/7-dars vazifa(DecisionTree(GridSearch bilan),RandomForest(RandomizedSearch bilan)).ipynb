# DecisionTree

```python
import pandas as pd
df=pd.read_csv('personality_dataset.csv')
```

```python
df.info()
```

```python
def tozala(df):
    for col in df.columns:
        if df[col].isnull().any():
            if df[col].dtype=='object':
                df[col].fillna(df[col].mode()[0],inplace=True)
            else:
                df[col].fillna(df[col].mean(),inplace=True)
    return df
```

```python
df=tozala(df)
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()

def encod_qil(df):
    for col in df.columns:
        if df[col].dtype=='object':
            df[col]=encoder.fit_transform(df[col])
    return df
```

```python
df=encod_qil(df)
```

```python
df.info()
```

```python
x=df.drop('Personality',axis=1)
```

```python
y=df['Personality']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
dt_model=DecisionTreeClassifier()
```

```python
dt_model.fit(x_train,y_train)
```

```python
y_pred=dt_model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc1=accuracy_score(y_test,y_pred)
```

```python
print(f'DecisionTree aniqligi:{acc1*100}%ga teng')
```

# DecisionTree Grid Search Hyperparametr Tuning bilan

```python
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
dt_model = DecisionTreeClassifier(random_state=42)
```

```python
# Hyperparametr Tuning bilan
param_grid = {
    'max_depth': [3, 5, 10, None],
    'min_samples_split': [2, 5, 10],
    'criterion': ['gini', 'entropy']
}
```

```python
grid_search=GridSearchCV(estimator=dt_model, param_grid=param_grid, cv=5, scoring='accuracy')
```

```python
grid_search.fit(x,y)
```

```python
print("Best Parameters:", grid_search.best_params_)
```

```python
best_model=grid_search.best_estimator_
```

```python
best_model
```

```python
print("Best Score:", grid_search.best_score_)
```

# RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
rf_model=RandomForestClassifier()
```

```python
rf_model
```

```python
rf_model.fit(x_train,y_train)
```

```python
y_pred=rf_model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc2=accuracy_score(y_test,y_pred)
```

# RandomForest RandomizedSearch Hyperparametr Tuning bilan

```python
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
```

```python
# Hyperparametr Tuning bilan
param={
    'n_estimators':[10,100,150],
    'max_depth': [3, 5, 10, None],
    'min_samples_split':[2,5,10]
}
```

```python
random_search=RandomizedSearchCV(rf_model,param_distributions=param,n_iter=10,cv=5,random_state=42,n_jobs=-1)
```

```python
random_search.fit(x_train,y_train)
```

```python
print("Best Parameters:", random_search.best_params_)
```

```python
best_model=random_search.best_estimator_
```

```python
best_model
```

```python
print("Best Score:", random_search.best_score_)
```
