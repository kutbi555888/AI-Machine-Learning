# Logistic Regression

```python
import pandas as pd
df=pd.read_csv('diabetes_prediction_dataset.csv')
```

```python
df.info()
```

```python
# missing valuelar yoq,lekin missing value kodini yozib ketamiz
def tozala(df):
    for col in df.columns:
        if df[col].isnull().any():
            if df[col].dtype=='object':
                df[col].fillna(df[col].mode()[0],inplace=True)
            else:
                df[col].fillna(df[col].mean(),inplace=True)
                
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()

# encoding
def encod_qil(df):
    for col in df.columns:
        if df[col].dtype=='object':
            df[col]=encoder.fit_transform(df[col])
    return df
```

```python
df=encod_qil(df)
```

```python
df.info()
```

```python
# Scaling
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()

def scaling_qil(df):
    num_col=df.select_dtypes(include=['int32','float64']).columns
    if 'diabetes' in num_col:
        num_col = num_col.drop('diabetes')  # Faqat bor bo‘lsa tushiramiz
    df[num_col]=scaler.fit_transform(df[num_col])
    return df
```

```python
df=scaling_qil(df)
```

```python
df.head()
```

# Training

```python
x=df.drop('diabetes',axis=1)
```

```python
y=df['diabetes']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
lc_model=LogisticRegression()
```

```python
lc_model
```

```python
lc_model.fit(x_train,y_train)
```

```python
y_pred=lc_model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc1=accuracy_score(y_test,y_pred)
```

```python
print(f'LogisticRegression aniqligi:{acc1*100}%ga teng')

#2. f'...' → bu f-string (formatlangan string)
# f harfi formatlash degani: matn ichida o‘zgaruvchilarni bevosita joylashtirish imkonini beradi.

#F-string ichida {} ichiga o‘zgaruvchi nomini yozsangiz, Python u o‘zgaruvchining qiymatini joylashtiradi.

# 3. 'LogisticRegression aniqligi:{acc1*100}%ga teng'
#Bu yerda matn va o‘zgaruvchi aralashtirilgan:

#'LogisticRegression aniqligi: → bu oddiy matn.

#{acc1*100} → bu f-string ichidagi hisoblash.

#acc1 — bu sizning oldin hisoblangan aniqlik (accuracy) qiymatingiz.

#acc1*100 — accuracy ni foizga aylantiryapti (masalan: 0.83 → 83.0).

#'%ga teng' — bu oddiy matn bo‘lib, foiz degan belgi ko‘rinadi.

# f'...'	Formatlangan matn
# {acc1*100}	acc1 qiymatini 100 ga ko‘paytiradi
```

# LogisticRegression K-cross validation bilan

```python
from sklearn.model_selection import cross_val_score,KFold
```

```python
kf1=KFold(n_splits=5,shuffle=True,random_state=42)
scores1=cross_val_score(lc_model,x,y,cv=kf1,scoring='accuracy')
print(scores1)
```

```python
print(scores1)
```

```python
import numpy as np
```

```python
print(np.mean(scores1).round(4))
```

```python
np.std(scores1)
```
